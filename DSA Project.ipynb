```python
unique_years = df['Year'].unique()

def assign_colors(num_colors):
    colors = plt.cm.tab10(range(num_colors))
    return colors

for year in unique_years:
    year_data = df[df['Year'] == year]

    murders = year_data[year_data['Offence'] == 'Murder']

    regions = ['Punjab', 'Sindh', 'KP', 'Balochistan', 'Islamabad', 'Railways', 'G.B', 'AJK']
    murder_cases = murders[regions].iloc[0]

    num_regions = len(regions)
    colors = assign_colors(num_regions)

    plt.figure(figsize=(10, 6))
    bars = plt.bar(regions, murder_cases, color=colors)

    plt.title(f'Number of Murders in Different Regions of Pakistan ({year})')
    plt.xlabel('Regions')
    plt.ylabel('Number of Murders')

    for i, v in enumerate(murder_cases):
        plt.text(i, v, str(v), ha='center', va='bottom')

    plt.show()
```

```python
unique_years = df['Year'].unique()

def assign_colors(num_colors):
    colors = plt.cm.tab10(range(num_colors))
    return colors

for year in unique_years:
    year_data = df[df['Year'] == year]

    Robbery = year_data[year_data['Offence'] == 'Robbery']

    regions = ['Punjab', 'Sindh', 'KP', 'Balochistan', 'Islamabad', 'Railways', 'G.B', 'AJK']
    Robbery_cases = Robbery[regions].iloc[0]
    
    num_regions = len(regions)
    colors = assign_colors(num_regions)

    plt.figure(figsize=(10, 6))
    bars = plt.bar(regions, Robbery_cases, color=colors)

    plt.title(f'Number of Robbery in Different Regions of Pakistan ({year})')
    plt.xlabel('Regions')
    plt.ylabel('Number of Robbery')

    for i, v in enumerate(Robbery_cases):
        plt.text(i, v, str(v), ha='center', va='bottom')

    plt.show()
```
